# Loading and Exploring Data

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df = pd.read_csv('Data/vgsales.csv')
df.head()
```

```python
# Check data types and missing data
df.info()
```

```python
# check for duplicate data
df.duplicated().sum()
```

```python
for col in df.select_dtypes('object'):
    print(df[col].value_counts())
```

```python
for col in df.select_dtypes('int64'):
    print(df[col].value_counts())
```

```python
for col in df.select_dtypes('float'):
    print(df[col].value_counts())
```

```python
#check for missing
df.isna().sum()
```

```python
#Locate all missings
df.loc[df.isna().any(axis=1)]
```

```python
#View missing year
df[df['Year'].isna()]
```

```python
#View missing publisher
df[df['Publisher'].isna()]
```

For all the missing data I am going to replace with missing so I do not miss out on important sales data from removing the games that are missing data. 

```python
# replace Nan with missing
df['Year'].fillna('Missing',inplace = True)
df['Publisher'].fillna('MIssing',inplace=True)
```

```python
#check fo rmissings again
df.isna().sum()
```

```python
dtypes = df.dtypes
str_cols = dtypes[dtypes=='object'].index
str_cols
```

```python
for col in str_cols:
    print(f'- Columns={col}')
    print(df[col].value_counts(dropna= False))
    print('\n\n')
```

```python
dtypes = df.dtypes
str_cols = dtypes[dtypes == 'float'].index
str_cols
```

```python
for float in str_cols:
    print(f'- Columns={float}')
    print(df[float].value_counts(dropna = False))
    print('\n\n')
    
    
```
